```python
import matplotlib
matplotlib.__version__
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as thetotal fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Change the plot size
plt.figure(figsize=(20, 10))
# Create the plot and add a lable for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create a lable for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Set the x limit between January through June
plt.xlim("Jan", "Jun")
# Create a title
plt.title('PyBer Fare by Month')
# Add the legend
plt.legend()
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
ax.set_title("This is my title")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.grid()
ax.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create lables for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create the title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create a horizontal plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
# Add title
plt.title("PyBer Fare by Month")
# Lable axes
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Add legend
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k", label="New York")
#ax.set_label()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
# Set variables, color and label
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
# Set axis labels
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Invert axis
ax.invert_yaxis()
# Set legend
ax.legend()
```
